```python
import numpy as np
import pandas as pd
```

```python
movies = pd.read_table('data_sets\movies.dat', delimiter='::', names=['MovieID', 'Title', 'Genres'])
movies
```

```python
ratings = pd.read_table('data_sets/ratings.dat', delimiter='::', names=['UserID', 'MovieID', 'Rating', 'Timestamp'])
ratings
```

```python
users = pd.read_table('data_sets/users.dat', delimiter='::', names=['UserID', 'Gender', 'Age', 'Occupation', 'Zip-code'])
users
```

```python
ratings_users = pd.merge(ratings, users)
ratings_users
```

```python
frame = pd.merge(ratings_users, movies)
frame
```

```python
piv_rating_gend = pd.pivot_table(frame, values='Rating', index='Title', columns='Gender')
piv_rating_gend
```

```python
piv_rating_age = pd.pivot_table(frame, values='Rating', columns='Age', index='Title')
piv_rating_age
```

```python
piv_rating_gend = piv_rating_gend.sort_values('F', ascending=False)
piv_rating_gend
```

```python
piv_rating_age = piv_rating_age.sort_values(max(frame['Age']), ascending=False)
piv_rating_age
```

```python
group_tr = frame[['Title', 'Rating']].groupby('Title').count()
group_tr
```

```python
tr_series = group_tr['Rating']
tr_series
```

```python
tr_series[tr_series >= 250]
```

```python
subtract = pd.Series(piv_rating_gend['F'] - piv_rating_gend['M'], index=piv_rating_gend.index)
subtract
```

```python
# Первые 15 фильмов, которым женская аудитория поставила высокие оценки, а мужская - низкие.
subtract.sort_values(ascending=False)[np.arange(15)]
```

```python
# Первые 15 фильмов, которым мужская аудитория поставила высокие оценки, а женская - низкие.
subtract.sort_values()[np.arange(15)]
```
